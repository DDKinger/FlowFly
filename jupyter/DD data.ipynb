```python
from openpyxl import Workbook, load_workbook
import itertools
import pandas as pd
import os
from tqdm import tqdm
```

```python
def load_file(fn):
    wb = load_workbook(filename=fn)
    sheet = wb['导出工作表']
    data = sheet.values
    cols = next(data)
    data = list(data)
    idx = [i for i, r  in enumerate(data)]
    df = pd.DataFrame(data, index=idx, columns=cols)
    df['RECORD_TIME'] = df['RECORD_TIME'].map(pd.Timestamp)
    df['POSITION'] = int(fn.strip('.xlsx'))
    return df
```

```python
files = [f for f in os.listdir() if f.endswith('.xlsx')]
```

```python
load_file(files[0])
```

```python
dfs = [load_file(f) for f in tqdm(files)]
```

```python
df = pd.concat(dfs, ignore_index=True)
```
